✅ Why Do We Do Feature Scaling?
1. To Ensure Fair Contribution of Features
Features with larger ranges can dominate distance-based models.

Example:

Age ranges from 0–100.

Income ranges from 0–100,000.

Without scaling, income dominates, making the model biased toward it.

Standardization :it is a very effective technique which rescales the feature value so that it is distributed with mean =0 and variance=1
X(new)=(X(i)-X(mean))/Std

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("../loan.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
ss.transform(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["x"])
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["ApplicantIncome_ss"])
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

Normalization:
it is a Scaling technique in which values are shifted and rescaled so that they end up raging between 0 and 1. it is also known as min-max scaling
X(new)=(X(i)-min(X))/(max(x)-min(x))
