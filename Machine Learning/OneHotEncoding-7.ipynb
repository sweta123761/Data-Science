One hot encoding is used when no. of data is less(like yes/no,urban/rural/semiurban)
it assigns values like 001,110 etc

```python
import pandas as pd
```

```python
dataset= pd.read_csv("../loan.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)  #categorical data hence we have used mode
```

```python
dataset.isnull().sum()
```

```python
dataset["Married"].fillna(dataset["Married"].mode()[0],inplace=True) 
```

```python
#one hot encoding
en_data = dataset[["Gender","Married"]]
pd.get_dummies(en_data)
```

                                 ---------------
We can also do it directly using |Scikit learn |
                                 ---------------

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe= OneHotEncoder()
ar=ohe.fit_transform(en_data).toarray()
ar
```

```python
pd.DataFrame(ar,columns=['Gender_Female','Gender_Male','Married_No','Married_Yes'])
```

Solution: drop='first'
By dropping the first category (e.g., "Red"), you get:

Color	Green	Blue
Red	0	0
Green	1	0
Blue	0	1

The model infers the dropped category ("Red") when both Green and Blue are 0.



```python
ohe= OneHotEncoder(drop="first")
ar=ohe.fit_transform(en_data).toarray()
ar
```

```python
pd.DataFrame(ar,columns=['Gender_Male','Married_Yes'])
```
