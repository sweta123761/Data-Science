jab data in normal distribution form me hota hai.aur sume jo points bahut jyada door chale jaate hai aur tail ko lamba kar dete hai unko ham outliers bolte hai.

Jab data ek given range ke bahar chala jaye to ham usko outliers bolte hai

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

```

```python
dataset = pd.read_csv("../loan.csv")
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
sns.boxplot(x= "CoapplicantIncome",data=dataset) #find outliers using boxplot
plt.show()
```

```python
sns.distplot(dataset["CoapplicantIncome"])#another way to see outliers
plt.show()
```

```python
#how to solve this outlier problem
#IQR method(inter quantile range
dataset.shape
```

```python
q1=dataset["CoapplicantIncome"].quantile(0.25)
q3=dataset["CoapplicantIncome"].quantile(0.75)
IQR=q3-q1
min_range = q1-(1.5*IQR)
max_range = q3+(1.5*IQR)
min_range,max_range
```

```python
new_dataset = dataset[dataset["CoapplicantIncome"]<=max_range]
new_dataset.shape
```

```python
sns.boxplot(x= "CoapplicantIncome",data=new_dataset) 
plt.show()
```

Outlier removal using Z-Score
in normal graph data outside 3rd Standard deviation is treated as outliers
Z=(x-u)/std
u=Mean,Std=Standard Deviation

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 

```

```python
dataset = pd.read_csv("../loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.boxplot(x="CoapplicantIncome",data=dataset)
```

```python
sns.distplot(dataset["CoapplicantIncome"])
```

```python
#we wourld remove outliers through 2 methods 1.Normal method ,2.Z Score Method
min_range= dataset["CoapplicantIncome"].mean()-(3*dataset["CoapplicantIncome"].std())
max_range= dataset["CoapplicantIncome"].mean()+(3*dataset["CoapplicantIncome"].std())
min_range,max_range
```

```python
new_dataset=dataset[dataset["CoapplicantIncome"]<=max_range]
new_dataset.shape
```

```python
sns.boxplot(x="CoapplicantIncome",data=new_dataset)
```

```python
#-----------Z Score way----------
z_score = (dataset["CoapplicantIncome"]-dataset["CoapplicantIncome"].mean())/(dataset["CoapplicantIncome"].std())
z_score
```

```python
dataset["z_score"]=z_score
dataset
```

```python
dataset[dataset["z_score"]<3]
```
