jab bhi hame koi population data le(Normal/Not Normal)-aur usme hame small samples divide karde -aur phir us sample ke mean nikalkar graph par plot kare to hamesha normal distribution graph banega--this is called central limit theorum
aur jam bhi ham vo jo small samples ke mean ka total mean nikalte hai to vo total mean ke equal aata hai
and variance would be equal to standard error

there are few assumptions of central limit theorum :
1.the sample data should be selected randomly
2.there should not be any multicollinearity in the sample data (one sample does not influence other samples)
3.the sample size should not be more than 10% of the population .genrally a sample greater than 30 is considered good

```python
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
pop_data=[np.random.randint(10,100) for i in range(1000)]
pop_table= pd.DataFrame({"pop_data":pop_data})
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="pop_data",data=pop_table)
plt.show()
```

```python
sample_mean = []
for no_sample in range(50):
    sample_data = []
    for data in  range(500):
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))
sample_M=pd.DataFrame({"Sample_Mean":sample_mean})
plt.figure(figsize=(3,2))
sns.kdeplot(x="Sample_Mean",data=sample_M)
plt.show()
```
